Scraper para obter os links para download dos arquivos de áudio do podcast "This American Life" (https://www.thisamericanlife.org/).

```python
#links-processed internal
#tal.fm/
from bs4 import BeautifulSoup as bsp
import requests
import re
import pandas as pd
import numpy as np
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
```

```python
#scrape the website
base_url = 'https://tal.fm/'
i = 0
link_dict = {}
while True:
    i += 1
    print(i)
    url = base_url + str(i)
    response = requests.get(url)
    #if 404, break
    if response.status_code == 404:
        break

    #find link with class = "links-processed internal"
    soup = bsp(response.text, 'html.parser')
    links = soup.find_all('li', class_ = "download")
    if len(links) != 0:
        link = links[0].find('a')['href']
        #append links to list
        link_dict[i] = link

#convert to dataframe
df = pd.DataFrame.from_dict(link_dict, orient = 'index', columns = ['link'])
df['episode'] = df.index
df = df.reset_index(drop = True)
df['episode'] = df['episode'].astype(int)

#save to csv file in /data
df.to_csv('data/tal_links.csv', index = False)
```
